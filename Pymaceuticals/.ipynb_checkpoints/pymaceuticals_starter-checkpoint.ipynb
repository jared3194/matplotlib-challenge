## Observations and Insights
1. Capomulin and Ramicane seem to be the most effective treatments with the lowest average tumor volumes and lowest volume variability per the standard deviation.

2. The above observation is further reinforced by the chart that plots the total number of mice tested per each drug treatement. All drugs except Capomulin and Ramicane were tested on a lower population of mice yet they still produced higher tumor volume results. Capmulin and Ramicane maintained the lowest averages despite having a larger test population.

3. There is a strong positive correlation between between mouse weight and average tumor volume. This could mean that the drugs become less effective as the mice increase in body weight which would indicate that larger patients should receive larger doses of a drug to reduce tumor volume.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_df = pd.read_csv(mouse_metadata_path)
results_df = pd.read_csv(study_results_path)

# Combine the data into a single dataset

df = pd.merge(mouse_df, results_df, on="Mouse ID")

# Display the data table for preview
df
```

```python
# Checking the number of mice.
grouped_df = df.groupby(["Mouse ID"])
print(len(grouped_df))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# duplicates = df.duplicated(subset = ['Mouse ID', 'Timepoint'], keep = False)
duplicates = df.loc[df.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicateRowsDF = df.loc[df['Mouse ID'] == 'g989']
duplicateRowsDF
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# sort dataframe by timepoint ascending and reset index
clean_df = df[df['Mouse ID'].isin(duplicates)==False]
clean_sorted = clean_df.sort_values(by=['Timepoint'], ascending = True)
final_df = clean_sorted.reset_index()
del final_df['index']
# Rearrange colums
titles = list(final_df.columns)
titles[1],titles[2],titles[3],titles[4],titles[5],titles[6],titles[7] = titles[5], titles[6], titles[7],titles[1], titles[2], titles[3], titles[4]
final_df = final_df[titles]
final_df
final_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
grouped_final_df = final_df.groupby(['Mouse ID'])
number_mice = len(grouped_final_df)
print(number_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

group_final = final_df.groupby(['Drug Regimen'])
tumor_values = group_final['Tumor Volume (mm3)']


mean_tumor = tumor_values.mean()
median_tumor = tumor_values.median()
variance_tumor = tumor_values.var()
standard_deviation_tumor = tumor_values.std()
sem_tumor = tumor_values.sem()


sum_stats = pd.DataFrame({
                         "Mean Tumor Volume (mm3)": mean_tumor,
                         "Median Tumor Volume (mm3)": median_tumor,
                         "Variance Tumor Volume (mm3)": variance_tumor,
                         "Standard Deviation Tumor Volume (mm3)": standard_deviation_tumor,
                         "SEM Tumor Volume (mm3)": sem_tumor})
sum_stats


```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
tumor_values.agg(['mean', 'median', 'var', 'std', 'sem'])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

final_final_tumor_vol = df[df["Drug Regimen"].isin(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])]
final_final_tumor_vol = final_final_tumor_vol.sort_values(["Timepoint"], ascending=True)


final_final_tumor_vol = final_final_tumor_vol[["Drug Regimen", "Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]

final_final_tumor_vol
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in drug_list:
    quartiles = final_final_tumor_vol.loc[final_final_tumor_vol['Drug Regimen'] == drug, 'Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    print(f"The lower quartile of {drug} is: {lowerq}")
    print(f"The upper quartile of {drug} is: {upperq}")
    print(f"The interquartile range of {drug} is: {iqr}")
    print(f"The the median of {drug} is: {quartiles[0.5]} ")
   
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"{drug} Values below {lower_bound} could be outliers.")
    print(f"{drug} Values above {upper_bound} could be outliers.")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Create empty list to fill with tumor vol data (for plotting)
#Group data by Drug Regimen and Mouse ID to capture Last Tumor Measurement
final_final_tumor_vol_sort = final_final_tumor_vol.groupby(['Drug Regimen', 'Mouse ID']).last()['Tumor Volume (mm3)']

# Turn retrieved data into dataframe to easily manipulate
box_df = final_final_tumor_vol_sort.to_frame()
box_df

# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
final_df = box_df.reset_index()
tumor_lists = final_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list)
tumor_list_df = pd.DataFrame(tumor_lists)
tumor_list_df = tumor_list_df.reindex(drug_list)
tumor_vols = [vol for vol in tumor_list_df['Tumor Volume (mm3)']]
plt.boxplot(tumor_vols, labels=drug_list)
plt.ylim(10, 80)
plt.ylabel("Final Tumor Volume (mm3)")
plt.title('Final Tumor Volume per Drug Regimen')
plt.show()

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
grp_capomulin_df = Capomulin_df.groupby('Mouse ID').mean()

marker_size= 25
plt.scatter(grp_capomulin_df['Weight (g)'],grp_capomulin_df['Tumor Volume (mm3)'], color="blue",)
plt.title('Mouse Weight Versus Average Tumor Volume',fontsize = 15)
plt.xlabel('Weight (g)',fontsize =10)
plt.ylabel('Averag Tumor Volume (mm3)',fontsize =10)


# plt.savefig("../Images/scatterplot.png", bbox_inches = "tight")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = round(st.pearsonr(grp_capomulin_df['Weight (g)'],grp_capomulin_df['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {correlation}")


x_values = grp_capomulin_df['Weight (g)']
y_values = grp_capomulin_df['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept


line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Add regression line to scatter plot

plt.scatter(grp_capomulin_df['Weight (g)'],grp_capomulin_df['Tumor Volume (mm3)'], color="blue",)
plt.plot(x_values,regress_values,"r-")
plt.title('Mouse Weight Versus Average Tumor Volume',fontsize = 12)
plt.xlabel('Weight (g)',fontsize =10)
plt.ylabel('Averag Tumor Volume (mm3)',fontsize =10)

plt.show()

```
