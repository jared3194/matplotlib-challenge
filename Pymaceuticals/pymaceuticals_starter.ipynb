## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_df = pd.read_csv(mouse_metadata_path)
results_df = pd.read_csv(study_results_path)

# Combine the data into a single dataset

df = pd.merge(mouse_df, results_df, on="Mouse ID")

# Display the data table for preview
df
```

```python
# Checking the number of mice.
grouped_df = df.groupby(["Mouse ID"])
print(len(grouped_df))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# duplicates = df.duplicated(subset = ['Mouse ID', 'Timepoint'], keep = False)
duplicates = df.loc[df.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicateRowsDF = df.loc[df['Mouse ID'] == 'g989']
duplicateRowsDF
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# sort dataframe by timepoint ascending and reset index
clean_df = df[df['Mouse ID'].isin(duplicates)==False]
clean_sorted = clean_df.sort_values(by=['Timepoint'], ascending = True)
final_df = clean_sorted.reset_index()
del final_df['index']
# Rearrange colums
titles = list(final_df.columns)
titles[1],titles[2],titles[3],titles[4],titles[5],titles[6],titles[7] = titles[5], titles[6], titles[7],titles[1], titles[2], titles[3], titles[4]
final_df = final_df[titles]
final_df
final_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
grouped_final_df = final_df.groupby(['Mouse ID'])
number_mice = len(grouped_final_df)
print(number_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

group_final = final_df.groupby(['Drug Regimen'])
tumor_values = group_final['Tumor Volume (mm3)']


mean_tumor = tumor_values.mean()
median_tumor = tumor_values.median()
variance_tumor = tumor_values.var()
standard_deviation_tumor = tumor_values.std()
sem_tumor = tumor_values.sem()


sum_stats = pd.DataFrame({
                         "Mean Tumor Volume (mm3)": mean_tumor,
                         "Median Tumor Volume (mm3)": median_tumor,
                         "Variance Tumor Volume (mm3)": variance_tumor,
                         "Standard Deviation Tumor Volume (mm3)": standard_deviation_tumor,
                         "SEM Tumor Volume (mm3)": sem_tumor})
sum_stats


```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
tumor_values.agg(['mean', 'median', 'var', 'std', 'sem'])
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.
mouse_per_drug = group_final['Mouse ID'].count().sort_values(ascending=False)

mouse_plot = mouse_per_drug.plot(kind = 'bar', x = 'Drug Regimen', y = 'Mouse ID', color = 'b', title = "Number of Measurements Per Drug Regimen", ylabel = "Number of Unique Mice Tested")



```

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pyplot.
x_axis = np.arange(len(mouse_per_drug))
tick_locations = [value for value in x_axis]
drug_names = df["Drug Regimen"].unique()
plt.xticks(tick_locations, drug_names, rotation = 90)
plt.bar(x_axis, mouse_per_drug, color = 'b', alpha = 1, align = 'center')

plt.title("Number of Measurements Per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Unique Mice Tested")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
gender_df = final_df.groupby('Sex')
gender_count = gender_df["Mouse ID"].count().sort_values(ascending = False)

pie_plot = gender_count.plot(kind = 'pie', y = 'Mouse ID', title = "Distribution of Female vs Male Mice", ylabel = 'Sex', autopct='%1.1f%%')
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
sex = ['Male', 'Female']

count = gender_count

colors = ["blue", "orange"]

plt.pie(count, labels= sex, colors=colors,
        autopct="%1.1f%%", startangle = 0)

plt.title("Distribution of Female vs Male Mice")
plt.ylabel("Sex")
```
